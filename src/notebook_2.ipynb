## Feature Importance Analysis (Re-Done)

```python
import numpy as np
from sklearn.decomposition import PCA

# Extract the genre embeddings
genre_embeddings = np.stack(df_loaded['genre_embeddings'].values)

# Apply PCA to reduce dimensions, e.g., to 50 dimensions
pca = PCA(n_components=50)
embeddings_reduced = pca.fit_transform(genre_embeddings)

# Convert reduced embeddings back to a DataFrame and concatenate with the original DataFrame
embeddings_df_pca = pd.DataFrame(embeddings_reduced, columns=[f'PCA{i+1}' for i in range(embeddings_reduced.shape[1])])

# Reset the index of both DataFrames to ensure alignment
df_reset = df_loaded.reset_index(drop=True)
embeddings_df_pca_reset = embeddings_df_pca.reset_index(drop=True)

# Join the DataFrames
df_pca = pd.concat([df_reset, embeddings_df_pca_reset], axis=1)

# Select features and target
pca_features = [f'PCA{i+1}' for i in range(embeddings_reduced.shape[1])]
original_features_pca = ['duration_ms', 'explicit', 'danceability', 'energy', 'key', 'loudness', 'mode', 
                         'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 
                         'time_signature', 'decade_encoded', 'followers']
features_pca = original_features_pca + pca_features
X_pca = df_pca[features_pca]
y_pca = df_pca['hit_category_encoded']
```

```python
from sklearn.ensemble import RandomForestClassifier
# Initialize and train the Random Forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_pca, y_pca)

# Get feature importances
importances = rf_model.feature_importances_
feature_importances = pd.DataFrame({'Feature': X_pca.columns, 'Importance': importances})

# Sort by importance
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Plot feature importances
top_20_features = feature_importances.sort_values(by='Importance', ascending=False).head(20)

plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=top_20_features)
plt.title('Feature Importances')
plt.show()
```

```python
# Build the neural network for PCA data
model_pca = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_pca_scaled.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])

# Compile the model
model_pca.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history_pca = model_pca.fit(X_train_pca_scaled, y_train_pca_categorical,
                            epochs=30, batch_size=64,
                            validation_data=(X_val_pca_scaled, y_val_pca_categorical))
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score
# Function to test and evaluate the model
def testingModel(model, X, y):
    """ 
    Function to test and evaluate the model
    Parameters:
        model: trained model
        X: input features
        y: target variable
    """
    y_pred = np.argmax(model.predict(X), axis=1)
    accuracy = accuracy_score(np.argmax(y, axis=1), y_pred)
    report = classification_report(np.argmax(y, axis=1), y_pred)
    f1_macro = f1_score(np.argmax(y, axis=1), y_pred, average='macro')
    return accuracy, report, f1_macro

# Evaluate the model on the validation set
acc_val_pca_nn, report_val_pca_nn, f1_val_pca_nn = testingModel(model_pca, X_val_pca_scaled, y_val_pca_categorical)

print(f"Validation Accuracy with PCA: {acc_val_pca_nn}")
print(f"Validation F1 Macro Score with PCA: {f1_val_pca_nn}")
print(f"Classification Report (Validation with PCA):\n{report_val_pca_nn}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, accuracy_score, f1_score

# Define a simple neural network model
def create_simple_model(input_shape, num_classes):
    """
    Creates a simple neural network model with two hidden layers
    Parameters:
        input_shape (int): The number of features
        num_classes (int): The number of classes
    """
    model = Sequential()
    model.add(Dense(64, activation='relu', input_shape=(input_shape,)))
    model.add(Dropout(0.3))
    model.add(Dense(32, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(num_classes, activation='softmax'))
    
    optimizer = Adam(learning_rate=0.001)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Build and train the simple model
simple_model = create_simple_model(X_train_pca_smote_scaled.shape[1], num_classes)
history_simple = simple_model.fit(X_train_pca_smote_scaled, y_train_pca_smote_categorical,
                                  epochs=30, batch_size=64,
                                  validation_data=(X_val_pca_scaled, y_val_pca_categorical))

# Evaluate the simple model
def evaluate_model(model, X, y):
    """
    Evaluates the model on the given data
    Parameters:
        model: trained model
        X: input features
        y: target variable
    """
    y_pred = np.argmax(model.predict(X), axis=1)
    accuracy = accuracy_score(np.argmax(y, axis=1), y_pred)
    report = classification_report(np.argmax(y, axis=1), y_pred)
    f1_macro = f1_score(np.argmax(y, axis=1), y_pred, average='macro')
    return accuracy, report, f1_macro

acc_val_simple, report_val_simple, f1_val_simple = evaluate_model(simple_model, X_val_pca_scaled, y_val_pca_categorical)
print(f"Validation Accuracy (Simple NN): {acc_val_simple}")
print(f"Validation F1 Macro Score (Simple NN): {f1_val_simple}")
print(f"Classification Report (Validation - Simple NN):\n{report_val_simple}")

acc_test_simple, report_test_simple, f1_test_simple = evaluate_model(simple_model, X_test_pca_scaled, y_test_pca_categorical)
print(f"Test Accuracy (Simple NN): {acc_test_simple}")
print(f"Test F1 Macro Score (Simple NN): {f1_test_simple}")
print(f"Classification Report (Test - Simple NN):\n{report_test_simple}")
```

```python
# Define a deeper neural network model
def create_deep_model(input_shape, num_classes):
    """
    Creates a deeper neural network model with three hidden layers
    Parameters:
        input_shape (int): The number of features
        num_classes (int): The number of classes
    """
    model = Sequential()
    model.add(Dense(128, activation='relu', input_shape=(input_shape,)))
    model.add(Dropout(0.3))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(32, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(num_classes, activation='softmax'))
    
    optimizer = Adam(learning_rate=0.001)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Build and train the deep model
deep_model = create_deep_model(X_train_pca_smote_scaled.shape[1], num_classes)
history_deep = deep_model.fit(X_train_pca_smote_scaled, y_train_pca_smote_categorical,
                              epochs=30, batch_size=64,
                              validation_data=(X_val_pca_scaled, y_val_pca_categorical))

# Evaluate the deep model
acc_val_deep, report_val_deep, f1_val_deep = evaluate_model(deep_model, X_val_pca_scaled, y_val_pca_categorical)
print(f"Validation Accuracy (Deep NN): {acc_val_deep}")
print(f"Validation F1 Macro Score (Deep NN): {f1_val_deep}")
print(f"Classification Report (Validation - Deep NN):\n{report_val_deep}")

acc_test_deep, report_test_deep, f1_test_deep = evaluate_model(deep_model, X_test_pca_scaled, y_test_pca_categorical)
print(f"Test Accuracy (Deep NN): {acc_test_deep}")
print(f"Test F1 Macro Score (Deep NN): {f1_test_deep}")
print(f"Classification Report (Test - Deep NN):\n{report_test_deep}")
```

## KNN Reduced Features (Continued from previous notebook)

```python
import numpy as np

# Define a correlation threshold
correlation_threshold = 0.7

knn_df = df_loaded.copy()

knn_features = ['duration_ms', 'explicit', 'danceability', 'energy', 'key', 'loudness', 'mode', 
                         'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 
                         'time_signature', 'decade_encoded', 'followers']
# Calculate the correlation matrix
corr_matrix = knn_df[knn_features].corr().abs()

# Select the upper triangle of the correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool_))

# Find features with absolute correlation greater than the threshold
to_drop = [column for column in upper.columns if any(upper[column] > correlation_threshold)]

# Drop the highly correlated features
knn_reduced = knn_df.drop(columns=to_drop)

print("Features to drop due to high correlation:", to_drop)
print("Remaining features:", knn_reduced.columns)
```

```python
features_pca_reduced = [feature for feature in original_features_pca if feature not in to_drop]
features_pca_reduced
```
